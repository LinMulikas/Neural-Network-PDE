```python
from Poisson import *
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D

net = PDENN(
    input_size=2, output_size=1, hidden_depth=10, hidden_size=12, 
    lr = 1e-2)

best_model = Poisson((0, 1), (0, 1), 10000, True, net=net)

%matplotlib inline
best_model.drawReal(500, 'coolwarm')
```

```python
best_model.net.optim = best_model.net.adam
```
